```python
import subprocess

def check_ffmpeg():
    try:
        result = subprocess.run(['ffmpeg', '-version'], stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
        if result.returncode == 0:
            print("FFmpeg is installed.")
            print(result.stdout.split('\n')[0])
        else:
            print("FFmpeg is not installed.")
            print(result.stderr)
    except FileNotFoundError:
        print("FFmpeg is not installed or not found in the system PATH.")

# Run the check
check_ffmpeg()
```

```python
import os
import subprocess

# Directories containing the raw audio files
audio_dirs = ['extracted_audio1', 'extracted_audio2']

# Directories to save preprocessed audio
processed_audio_dirs = ['processed_audio1', 'processed_audio2']

# Create directories if they don't exist
for directory in processed_audio_dirs:
    os.makedirs(directory, exist_ok=True)

# Function to preprocess audio
def preprocess_audio(input_audio_path, output_audio_path):
    # Add your preprocessing steps here
    # For example, you can use ffmpeg to resample, denoise, or apply other effects
    subprocess.run(['ffmpeg', '-i', input_audio_path, '-ar', '44100', '-ac', '1', output_audio_path])

# Preprocess audio files from each directory
for i, (audio_dir, processed_audio_dir) in enumerate(zip(audio_dirs, processed_audio_dirs), start=1):
    for audio_file in os.listdir(audio_dir):
        if audio_file.endswith('.wav'):
            input_audio_path = os.path.join(audio_dir, audio_file)
            output_audio_path = os.path.join(processed_audio_dir, f'processed_audio_{i}_{audio_file}')
            preprocess_audio(input_audio_path, output_audio_path)
            print(f"Preprocessed audio saved: {output_audio_path}")
```

```python
import os
import numpy as np
import librosa

# Directories containing the preprocessed audio files
preprocessed_audio_dirs = ['processed_audio1', 'processed_audio2']

# Directories to save extracted features
features_dirs = ['extracted_audio1', 'extracted_audio2']

# Create directories if they don't exist
for directory in features_dirs:
    os.makedirs(directory, exist_ok=True)

# Function to extract features from audio
def extract_features(audio_path):
    # Load the audio file
    y, sr = librosa.load(audio_path, sr=None)
    # Extract MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    # Extract chroma features
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    return mfccs, chroma

# Extract features from audio files in each directory
for i, (audio_dir, features_dir) in enumerate(zip(preprocessed_audio_dirs, features_dirs), start=1):
    for audio_file in os.listdir(audio_dir):
        if audio_file.endswith('.wav'):
            audio_path = os.path.join(audio_dir, audio_file)
            mfccs, chroma = extract_features(audio_path)
            features_path = os.path.join(features_dir, f'features_{i}_{os.path.splitext(audio_file)[0]}.npz')
            np.savez(features_path, mfccs=mfccs, chroma=chroma)
            print(f"Features extracted and saved: {features_path}")
```

```python
import os
from pydub import AudioSegment, silence

# Directories containing the preprocessed audio files
preprocessed_audio_dirs = ['processed_audio1', 'processed_audio2']

# Directories to save segmented audio
segmented_audio_dirs = ['segmented_audio1', 'segmented_audio2']

# Create directories if they don't exist
for directory in segmented_audio_dirs:
    os.makedirs(directory, exist_ok=True)

# Function to segment audio based on silence
def segment_audio(audio_path, min_silence_len=500, silence_thresh=-40):
    # Load audio file
    audio_segment = AudioSegment.from_file(audio_path)
    
    # Split audio into segments based on silence
    segments = silence.split_on_silence(audio_segment, min_silence_len=min_silence_len, silence_thresh=silence_thresh)
    
    return segments

# Segment audio files and save them to corresponding directories
for i, (audio_dir, segmented_audio_dir) in enumerate(zip(preprocessed_audio_dirs, segmented_audio_dirs), start=1):
    for audio_file in os.listdir(audio_dir):
        if audio_file.endswith('.wav'):
            audio_path = os.path.join(audio_dir, audio_file)
            
            # Segment audio
            segments = segment_audio(audio_path)
            
            # Save segmented audio files
            for j, segment in enumerate(segments):
                segment_path = os.path.join(segmented_audio_dir, f"{os.path.splitext(audio_file)[0]}_segment{j}.wav")
                segment.export(segment_path, format='wav')
                
                print(f"Segment {j+1} saved: {segment_path}")
```

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns

# Directories containing the segmented audio files
segmented_audio_dirs = ['segmented_audio1', 'segmented_audio2']

# Define the pitch ranges and corresponding states
pitch_ranges = {
    'low': (50, 150),
    'medium': (151, 300),
    'high': (301, 500)
}

# Function to determine pitch state based on pitch value
def get_pitch_state(pitch):
    for state, (low, high) in pitch_ranges.items():
        if low <= pitch <= high:
            return state
    return None

# Function to compute pitch from audio segment
def compute_pitch(audio_path):
    y, sr = librosa.load(audio_path, sr=None)
    pitches, magnitudes = librosa.core.piptrack(y=y, sr=sr)
    pitch = []
    for t in range(pitches.shape[1]):
        index = magnitudes[:, t].argmax()
        pitch.append(pitches[index, t])
    pitch = np.array(pitch)
    pitch = pitch[pitch > 0]
    mean_pitch = np.mean(pitch)
    return mean_pitch

# Function to construct transition matrix based on pitch states
def construct_transition_matrix_and_collect_pitches(segmented_audio_dir):
    # Initialize transition matrix
    states = list(pitch_ranges.keys())
    num_states = len(states)
    transition_matrix = np.zeros((num_states, num_states))
    prev_pitch_state = None
    pitch_states = []

    # Process each segmented audio file
    for segment_file in sorted(os.listdir(segmented_audio_dir)):
        if segment_file.endswith('.wav'):
            segment_path = os.path.join(segmented_audio_dir, segment_file)
            pitch = compute_pitch(segment_path)
            pitch_state = get_pitch_state(pitch)
            if pitch_state is None:
                continue

            pitch_states.append(pitch_state)

            # Record transitions
            if prev_pitch_state is not None:
                transition_matrix[states.index(prev_pitch_state), states.index(pitch_state)] += 1
            prev_pitch_state = pitch_state

    # Normalize transition matrix
    transition_matrix = transition_matrix / np.maximum(transition_matrix.sum(axis=1, keepdims=True), 1)

    return transition_matrix, pitch_states

# Function to plot transition matrix heatmap
def plot_transition_matrix_heatmap(transition_matrix, title):
    states = list(pitch_ranges.keys())
    plt.figure(figsize=(8, 6))
    sns.heatmap(transition_matrix, annot=True, fmt=".2f", cmap='viridis', xticklabels=states, yticklabels=states)
    plt.xlabel('Next State')
    plt.ylabel('Current State')
    plt.title(title)
    plt.show()

# Function to plot pitch states over time
def plot_pitch_states_over_time(pitch_states, title):
    states = list(pitch_ranges.keys())
    state_indices = [states.index(state) for state in pitch_states]
    
    plt.figure(figsize=(12, 4))
    plt.plot(state_indices, marker='o', linestyle='-', color='b')
    plt.yticks(range(len(states)), states)
    plt.xlabel('Segment Index')
    plt.ylabel('Pitch State')
    plt.title(title)
    plt.grid(True)
    plt.show()

# Construct and plot transition matrices and pitch states for each directory
for i, segmented_audio_dir in enumerate(segmented_audio_dirs, start=1):
    print(f"Constructing transition matrix for directory {segmented_audio_dir}...")
    transition_matrix, pitch_states = construct_transition_matrix_and_collect_pitches(segmented_audio_dir)
    print(f"Transition Matrix for directory {segmented_audio_dir}:")
    print(transition_matrix)
    # Plot the transition matrix heatmap
    plot_transition_matrix_heatmap(transition_matrix, f'Transition Matrix Heatmap for directory {segmented_audio_dir}')
    # Plot the pitch states over time
    plot_pitch_states_over_time(pitch_states, f'Pitch States Over Time for directory {segmented_audio_dir}')
    # Optionally, save the transition matrix to a file
    np.save(f'transition_matrix_{i}.npy', transition_matrix)
```
